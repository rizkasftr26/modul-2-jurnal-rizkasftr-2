```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
import pandas as pd
```

```python
Image = plt.imread("senheScramble.jpg")
plt.title("senheScramble")
plt.imshow(Image,cmap="gray")
```

```python
tinggi, lebar = Image.shape
rotasi = np.zeros((tinggi,lebar)).astype(int)
sudut= np.deg2rad(70)
cos=np.cos(sudut)
sin=np.sin(sudut)
tinggitengah = int((tinggi-1)/2)
lebartengah = int((lebar-1)/2)
for i in range (rotasi.shape[0]):
    for j in range (rotasi.shape[1]):
        x=(i-tinggitengah)*cos+(j-lebartengah)*sin
        y=-(i-tinggitengah)*sin+(j-lebartengah)*cos
        x = int(x + tinggitengah)
        y = int(y + lebartengah)
        if x>=0 and y>=0 and x<rotasi.shape[0] and y<rotasi.shape[1] :
            rotasi[i][j]= Image[x][y]
plt.title("Hasil Rotasi")
plt.imshow(rotasi,cmap="gray")
```
